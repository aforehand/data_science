```python
# median encoding -- ordinal and categorical
sale_median = y_tr.median()

X_tr_med = X_tr.copy()
X_val_med = X_val.copy()
X_test_med = X_test.copy()

for col in cat_cols+ord_cols:
    meds = y_tr.groupby(X_tr[col]).median()
    encode_dict = meds.to_dict()
    keys_val_test = pd.concat([X_val[col], X_test[col]]).unique()
    for key in keys_val_test:
        if not key in encode_dict:
            encode_dict[key] = sale_median
    X_tr_med[col].replace(encode_dict, inplace=True)
    X_val_med[col].replace(encode_dict, inplace=True)
    X_test_med[col].replace(encode_dict, inplace=True)
```

```python
# mean encoding -- ordinal and categorical
sale_mean = y_tr.mean()

X_tr_mean = X_tr.copy()
X_val_mean = X_val.copy()
X_test_mean = X_test.copy()

for col in cat_cols+ord_cols:
    means = y_tr.groupby(X_tr[col]).mean()
    encode_dict = means.to_dict()
    keys_val_test = pd.concat([X_val[col], X_test[col]]).unique()
    for key in keys_val_test:
        if not key in encode_dict:
            encode_dict[key] = sale_mean
    X_tr_mean[col].replace(encode_dict, inplace=True)
    X_val_mean[col].replace(encode_dict, inplace=True)
    X_test_mean[col].replace(encode_dict, inplace=True)
```

```python
# generate features giving the mean and median target values of the n-nearest neighbors
neighbors = NearestNeighbors(
    n_neighbors=500, metric='braycurtis', n_jobs=-1).fit(X_tr_med)

n_neighbors = [5, 10, 20, 50, 100, 200, 500]

meds_tr = {}; meds_val = {}; meds_test = {};
means_tr = {}; means_val = {}; means_test = {};
for i in n_neighbors:
    
    neighbors_tr_med = neighbors.kneighbors(X_tr_med, n_neighbors=i, return_distance=False)
    neighbors_val_med = neighbors.kneighbors(X_val_med, n_neighbors=i, return_distance=False)
    neighbors_test_med = neighbors.kneighbors(X_test_med, n_neighbors=i, return_distance=False)
    
    meds_tr['meds_{}'.format(i)] =  [y_tr.iloc[neighbors_tr_med[j]].median() for j in range(neighbors_tr_med.shape[0])]
    meds_val['meds_{}'.format(i)] =  [y_tr.iloc[neighbors_val_med[j]].median() for j in range(neighbors_val_med.shape[0])]
    meds_test['meds_{}'.format(i)] = [y_tr.iloc[neighbors_test_med[j]].median() for j in range(neighbors_test_med.shape[0])]
    
    neighbors_tr_mean = neighbors.kneighbors(X_tr_mean, n_neighbors=i, return_distance=False)
    neighbors_val_mean = neighbors.kneighbors(X_val_mean, n_neighbors=i, return_distance=False)
    neighbors_test_mean = neighbors.kneighbors(X_test_mean, n_neighbors=i, return_distance=False)
    
    means_tr['means_{}'.format(i)] =  [y_tr.iloc[neighbors_tr_mean[j]].mean() for j in range(neighbors_tr_mean.shape[0])]
    means_val['means_{}'.format(i)] =  [y_tr.iloc[neighbors_val_mean[j]].mean() for j in range(neighbors_val_mean.shape[0])]
    means_test['means_{}'.format(i)] = [y_tr.iloc[neighbors_test_mean[j]].mean() for j in range(neighbors_test_mean.shape[0])]


X_tr_med_nn = pd.concat([X_tr_med, pd.DataFrame(meds_tr, index=X_tr_med.index)], axis=1)
X_val_med_nn = pd.concat([X_val_med, pd.DataFrame(meds_val, index=X_val_med.index)], axis=1)
X_test_med_nn = pd.concat([X_test_med, pd.DataFrame(meds_test, index=X_test_med.index)], axis=1)

X_tr_mean_nn = pd.concat([X_tr_mean, pd.DataFrame(means_tr, index=X_tr_mean.index)], axis=1)
X_val_mean_nn = pd.concat([X_val_mean, pd.DataFrame(means_val, index=X_val_mean.index)], axis=1)
X_test_mean_nn = pd.concat([X_test_mean, pd.DataFrame(means_test, index=X_test_mean.index)], axis=1)
```

```python
# hyperparameter optimization
# returns the most frequent parameter values from the top 10 of 100 searches

def get_params(model, grid, X, y, searches=10, n_iter=10):
    results_list = []
    for i in range(1,searches+1):
        print(i)
        try:
            param_search = RandomizedSearchCV(estimator=model, 
                                              param_distributions=grid, 
                                              n_iter=n_iter, 
                                              n_jobs=-1).fit(X, y)
            results_list.append(param_search.cv_results_)
        except:
            pass

    results_dict = defaultdict(list)
    for d in results_list:
        for k,v in d.items():
            results_dict[k].extend(v)

    results_df = pd.DataFrame(results_dict)
    best_results = results_df.sort_values(by='mean_test_score', ascending=False)[:10]
    params_df = pd.DataFrame(list(best_results.params.values))

    modes = params_df.mode().iloc[0].to_dict()
    modes = {k: (int(v) if params_df[k].dtype == np.int64 else v) 
             for k, v in modes.items()}

    return modes
```

```python

class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
EPOCHS = 500
# # Shuffle the training set
# order = np.argsort(np.random.random(train_labels.shape))
# train_data = train_data[order]
# train_labels = train_labels[order]

def rms(prediction, labels):
    loss = tf.reduce_mean(tf.squared_difference(prediction, labels))
    return loss

def build_model(train_data):
    
    model = keras.Sequential([
    keras.layers.Dense(64, 
                       kernel_regularizer=keras.regularizers.l2(0.01),
                       activation=tf.nn.relu,
                       input_shape=(train_data.shape[1],)),
    keras.layers.Dense(64, 
                       kernel_regularizer=keras.regularizers.l2(0.01), 
                       activation=tf.nn.relu),
    keras.layers.Dense(1)
  ])

    optimizer = tf.train.RMSPropOptimizer(0.001)

    model.compile(loss=rms,
                optimizer=optimizer,
                metrics=[rms])
    return model

def plot_history(history):
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Root Mean Squared Logarithmic Error')
    plt.plot(history.epoch, np.array(history.history['loss']), label='Train Loss')
    plt.plot(history.epoch, np.array(history.history['val_loss']), label='Val Loss')
    plt.legend()
    plt.show()
```

```python
nn_med = build_model(X_tr_med_scaled)

history_med = nn_med.fit(X_tr_med_scaled, y_tr_scaled, epochs=EPOCHS,
                    validation_split=0.2, verbose=0,
                    callbacks=[PrintDot()])

pred_tr_scaled = nn_med.predict(X_tr_med_scaled).reshape(-1,1)
pred_val_scaled = nn_med.predict(X_val_med_scaled).reshape(-1,1)
pred_tr = np.exp(out_scaler.inverse_transform(pred_tr_scaled)) - 1
pred_val = np.exp(out_scaler.inverse_transform(pred_val_scaled)) - 1

print('train score: {}'.format(r2_score(pred_tr, y_tr)))
print('val score: {}'.format(r2_score(pred_val, y_val)))
```

```python
NFOLDS=5
SEED=None
def get_oof(model, X_train, y_train, X_test):
    ntrain = X_train.shape[0]
    ntest = X_test.shape[0]
    kf = KFold(n_splits=NFOLDS, random_state=SEED)

    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest))

    for i, (train_index, test_index) in enumerate(kf.split(X_train)):
        X_tr = X_train[train_index]
        y_tr = y_train[train_index]
        X_val = X_train[test_index]

        model.fit(X_tr, y_tr)

        oof_train[test_index] = model.predict(X_val).ravel()
        oof_test_skf[i, :] = model.predict(X_test).ravel()

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
```

```python
med_models = {'gbr_med': gbr_med, 'forest_med': forest_med, 'linreg_med': linreg_med}
mean_models = {'knn_mean': knn_mean, 'forest_mean': forest_mean}
per_models = {'nn_per': nn_per, 'ridge_per': ridge_per, 'gbm_per': gbm_per}

base_predictions_train = {}
base_predictions_test = {}

for label, model in med_models.items():
    train_oof, test_oof = get_oof(model, X_train_med_scaled, y_scaled, X_test_med_scaled)
    base_predictions_train[label] = train_oof.ravel()
    base_predictions_test[label] = test_oof.ravel()

for label, model in mean_models.items():
    train_oof, test_oof = get_oof(model, X_train_mean_scaled, y_scaled, X_test_mean_scaled)
    base_predictions_train[label] = train_oof.ravel()
    base_predictions_test[label] = test_oof.ravel()
    
for label, model in per_models.items():
    train_oof, test_oof = get_oof(model, X_train_per_scaled, y_scaled, X_test_per_scaled)
    base_predictions_train[label] = train_oof.ravel()
    base_predictions_test[label] = test_oof.ravel()
```
